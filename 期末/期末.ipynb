### 1. 国立大学の評価基準

#### (1) 大学の目的
　文部科学省よる学校教育法の第五十二条より，[大学の目的[1]](https://www.mext.go.jp/b_menu/shingi/chukyo/chukyo4/003/gijiroku/attach/1414357.htm)は，以下のように定められている．  
> 大学は、学術の中心として、広く知識を授けるとともに、深く専門の学芸を教授研究し、知的、道徳的及び応用的能力を展開させることを目的とする。  
  
　これに基づけば，大学に求められる役割は，以下の3つに分類・集約されると考えられる．  
- ➀ 教育：内部の人間に対して，広い知識，深い専門性，道徳（倫理）観，応用力を授ける役割
- ➁ 研究：学術の中心として，深く専門の学芸を研究し，新たな発見，未知に対する解明をする役割
- ➂ 社会貢献：獲得した学術・学芸を活用・応用し，社会に利潤や豊かさを与える役割

#### (2) 大学の目的に対する評価式
　前節のような大学の目的を受けて，それに対する評価式を考える．
　➀の内容に関して，”内部の人間に対して～を授ける役割（能力）”とは，1教員数あたりの学生数に比例すると考えられる．また，”広い知識，深い専門性”は，分野数とそのレベルの積で表すことができると考えられる．加えて，”道徳（倫理）観”については，それが欠如している場合は，何か事故や事件が起こるはずであると考えられ，目立った事件があれば減点する．逆に，”応用力”があれば，学生自ら何かに貢献していることがあると考えられるため，大学の表彰にしたがって，加点する．以上より，➀の教育に関するスコアは以下のように表現することができる．
$$S_{education} = \frac{学生数}{教員数} \times \left\{\sum_{i=1}^{分野数} (分野_i \times レベル_i) - 道徳・倫理的減点 + 応用的加点 \right\}$$  
　➁の内容に関して，”学術の中心として，深く専門の学芸を研究”とは，論文の被引用数によって表されると考えられる．この被引用数は，質の高い優秀な論文ほど被引用数が多くなると考えられ，論文数すなわち量が多いほど引用される確率が高くなると考えられるため，論文の質と量のバランスがうまく取れた指標であると考えられる．以上より，➁の研究に関するスコアを，以下のように表現することができる．
$$S_{study} = 合計被引用数 $$  
　➂の内容に関して，”獲得した学術・学芸を活用・応用し，社会に利潤や豊かさを与える役割（能力）”とは，共同研究によって生まれるであろう特許の数や，開発プロジェクトの件数によって表すことができると考えられる．以上より，➂の社会貢献に関するスコアを，以下のように表現することができる．
$$S_{social} = 特許数 + 開発プロジェクト数$$  

#### (3) 必要なデータの取得と総合的な評価式
　前節の評価式に含まれるデータの取得を試みる．各大学における学生数，教員数，特許数は，大学のホームページ[[2]](https://www.saitama-u.ac.jp/)[[3]](https://www.chiba-u.ac.jp/)[[4]](https://www.ynu.ac.jp/)より，各年度のデータが取得可能であるため，それを用いる．一方で，分野数やそのレベル，事件や事故，表彰，被引用数，開発プロジェクト数に関して，平等に比較可能である情報源が見つからなかった．特に➁の研究に関する評価指標については，被引用数やh-index，Impact factorなど客観的な指標を用いることを理想としたが，いずれのデータも見つからず，スクレイピングもうまくいかなかった．今回は，科研費の取得額を評価指標とする．したがって，今回は式を簡易化して，次式のように表す．

$$S_{all} = S_{education} + S_{study} + S_{social} = \frac{学生数}{教員数} + 科研費 + 特許数$$  

　加えて，各スコアの合計を取っているが，各スコアの影響を均等にするために標準化してから用いる．

#### (4) 欠損値の補間

### 2. さいたま市（埼玉大学）の現状

#### (1) 各種スコアの計算
　第1章の計算式に基づき，各種スコアおよびその合算値の計算を行う．また，標準化は3大学における諸値 $X$ の平均値 $\mu$ と標準偏差 $\sigma$ を用いて，次式のような $Z$ スコアを求めたものを用いる．
$$ Z = \frac{X - \mu}{\sigma} $$

```python
# 教育に関するスコア：S_educationの計算
df_su['S_education'] = df_su['student'] / df_su['professor']
df_tu['S_education'] = df_tu['student'] / df_tu['professor']
df_yu['S_education'] = df_yu['student'] / df_yu['professor']

# 研究に関するスコア：S_studyの計算
df_su['S_study'] = df_su['fund']
df_tu['S_study'] = df_tu['fund']
df_yu['S_study'] = df_yu['fund']

# 社会貢献に関するスコア：S_socialの計算
df_su['S_social'] = df_su['patent']
df_tu['S_social'] = df_tu['patent']
df_yu['S_social'] = df_yu['patent']


# 各スコアの平均と標準偏差を求める
df_scores = pd.concat([
    df_su[['year', 'S_education', 'S_study', 'S_social']].assign(university='Saitama University'),
    df_tu[['year', 'S_education', 'S_study', 'S_social']].assign(university='Chiba University'),
    df_yu[['year', 'S_education', 'S_study', 'S_social']].assign(university='Yokohama National University')
])
mean_education = df_scores['S_education'].mean()
std_education = df_scores['S_education'].std()
mean_study = df_scores['S_study'].mean()
std_study = df_scores['S_study'].std()
mean_social = df_scores['S_social'].mean()
std_social = df_scores['S_social'].std()

# 標準化されたスコアの追加
df_su["S'_education"] = (df_su['S_education'] - mean_education) / std_education
df_tu["S'_education"] = (df_tu['S_education'] - mean_education) / std_education
df_yu["S'_education"] = (df_yu['S_education'] - mean_education) / std_education

df_su["S'_study"] = (df_su['S_study'] - mean_study) / std_study
df_tu["S'_study"] = (df_tu['S_study'] - mean_study) / std_study
df_yu["S'_study"] = (df_yu['S_study'] - mean_study) / std_study

df_su["S'_social"] = (df_su['S_social'] - mean_social) / std_social
df_tu["S'_social"] = (df_tu['S_social'] - mean_social) / std_social
df_yu["S'_social"] = (df_yu['S_social'] - mean_social) / std_social

# スコアの合計値
df_su["S'_all"] = df_su["S'_education"] + df_su["S'_study"] + df_su["S'_social"]
df_tu["S'_all"] = df_tu["S'_education"] + df_tu["S'_study"] + df_tu["S'_social"]
df_yu["S'_all"] = df_yu["S'_education"] + df_yu["S'_study"] + df_yu["S'_social"]

su_sum = df_su.sum(numeric_only=True).to_frame().T
su_sum['year'] = 'sum'
df_su = pd.concat([df_su, su_sum], ignore_index=True)

tu_sum = df_tu.sum(numeric_only=True).to_frame().T
tu_sum['year'] = 'sum'
df_tu = pd.concat([df_tu, tu_sum], ignore_index=True)

yu_sum = df_yu.sum(numeric_only=True).to_frame().T
yu_sum['year'] = 'sum'
df_yu = pd.concat([df_yu, yu_sum], ignore_index=True)


display(df_su)
display(df_tu)
display(df_yu)
```

```python
# 教育に関するスコア：S_educationの描画
plt.figure()
plt.rcParams['font.family'] = 'Times New Roman'
plt.plot(df_su["year"][:9], df_su['S_education'][:9], label='Saitama University', marker='o')
plt.plot(df_tu["year"][:9], df_tu['S_education'][:9], label='Chiba University', marker='o')
plt.plot(df_yu["year"][:9], df_yu['S_education'][:9], label='Yokohama National University', marker='o')
plt.title('Trends in Education Scores')
plt.xlabel('Year')
plt.ylabel('Student-to-Teacher Ratio [-]')
plt.legend(bbox_to_anchor=(0.5, 1.2), loc='upper center', frameon=False, shadow=True, ncol=3, fancybox=False, edgecolor='black')
plt.grid(True)
plt.show()

# 研究に関するスコア：S_studyの描画
plt.figure()
plt.rcParams['font.family'] = 'Times New Roman'
plt.plot(df_su["year"][:9], df_su['S_study'][:9], label='Saitama University', marker='o')
plt.plot(df_tu["year"][:9], df_tu['S_study'][:9], label='Chiba University', marker='o')
plt.plot(df_yu["year"][:9], df_yu['S_study'][:9], label='Yokohama National University', marker='o')
plt.title('Trends in Study Scores')
plt.xlabel('Year')
plt.ylabel('Grants-in-Aid for Scientific Research [1000 yen]')
plt.legend(bbox_to_anchor=(0.5, 1.2), loc='upper center', frameon=False, shadow=True, ncol=3, fancybox=False, edgecolor='black')
plt.grid(True)
plt.show()

# 社会貢献に関するスコア：S_socialの描画
plt.figure()
plt.rcParams['font.family'] = 'Times New Roman'
plt.plot(df_su["year"][:9], df_su['S_social'][:9], label='Saitama University', marker='o')
plt.plot(df_tu["year"][:9], df_tu['S_social'][:9], label='Chiba University', marker='o')
plt.plot(df_yu["year"][:9], df_yu['S_social'][:9], label='Yokohama National University', marker='o')
plt.title('Trends in Social Contribution Scores')
plt.xlabel('Year')
plt.ylabel('Number of Patents Granted [-]')
plt.legend(bbox_to_anchor=(0.5, 1.2), loc='upper center', frameon=False, shadow=True, ncol=3, fancybox=False, edgecolor='black')
plt.grid(True)
plt.show()

# 合計スコアの描画
plt.figure()
plt.rcParams['font.family'] = 'Times New Roman'
plt.plot(df_su["year"][:9], df_su["S'_all"][:9], label='Saitama University', marker='o')
plt.plot(df_tu["year"][:9], df_tu["S'_all"][:9], label='Chiba University', marker='o')
plt.plot(df_yu["year"][:9], df_yu["S'_all"][:9], label='Yokohama National University', marker='o')
plt.title('Trends in Total Scores')
plt.xlabel('Year')
plt.ylabel('Standardized Total Score [-]')
plt.legend(bbox_to_anchor=(0.5, 1.2), loc='upper center', frameon=False, shadow=True, ncol=3, fancybox=False, edgecolor='black')
plt.grid(True)
plt.show()
```

#### (3) 比較から分かる埼玉大学の現状
　以外にも，$S_{all}$ の値が最も高くなったのは，埼玉大学であった．前章の指標に基づくスコアが最良であることから，埼玉大学は，大学の目的に対する達成度が高いと結論づけることができる．  
　各要素を確認すると，教員対学生比と特許取得件数のスコアが高く，科研費取得額のスコアが低いことが分かる．この結果から言えることは，教員1人が教育する学生数が多く，特許も年々増加傾向で取得し，社会に対して貢献できている一方で，研究費の取得が乏しく，研究の専門性・新規性・影響力がライバル市の大学と比較して劣っていることが言える．また，教員対学生比の多さは，広く学問を提供していることと，ある程度相関はあると考えれるが，教育の質に関しては，低下することが予測され，また，絶対的な学生数は他大よりも少なく，改善すべき点であると言える．加えて，データとして得られなかったが，埼玉大学の学部数は5個であり，千葉大学が10個，横浜国立大学が5個であり，学問の広さに関しても，違いがある事が分かる．学生数と学部数は，両方を考慮することで，教育の質を測ることができると考えられるため，今回の指標で学生数しか考慮できていない点は，欠陥であるといえる．  
　以上の内容から，さいたま市（埼玉大学）の良い点と悪い点は，以下のようにまとめられ，また，埼玉大学の現状は，以下のように総じることができる．  
##### ＜良い点＞
- 教員対学生比が高く，内部の教員は，広く知識を授けていると言える
- 特許取得件数が比較的多く，その数も年々増加しており，社会への貢献度が高い
- 前章の指標に基づくスコアが最良であり，大学の目的に対する達成度が高い
##### ＜悪い点＞
- 科研費取得額が比較的低く，客観的な研究力が認められていない
- 教育の質が比較的悪い恐れがある
- 学部数や絶対的な学生数が比較的少ない
##### ＜現状＞
> 他大とも比較した結果，大学の目的に対して忠実に成果を実現できていると言えるが，教員数および学部数の少なさに起因すると見られる絶対的な学生数の低さと，研究力が客観的に認められていないという2点が大きな課題である．

### 3. 埼玉大学の現状（課題）から考えられる将来のあるべき姿

#### (1) 課題に対するあるべき姿
　分析結果より，埼玉大学は研究力が客観的に認められるようになれば，今回の指標的に最良の結果となる．このことから，5～10年後の将来，埼玉大学が更に発展していくためには，研究力の増強が必要であることが分かる．したがって，さいたま市ないしは埼玉大学のあるべき姿は，以下のようにまとめることができる．  
> さいたま市は，埼玉大学に対して資金的援助，研究施設の建設，他大学との共同研究マッチングなどを促進し，埼玉大学はそれを受けて，”研究力のある大学”として生まれ変わる．

#### (2) あるべき姿を達成するための施策
　課題にもあるように，埼玉大学は教員数および学部数が少なく，それによって絶対的な学生数も少なくなっていると考えられる．したがって，外部から新たな分野に関わる教員と新たな学部の設置が求められる．現在，埼玉医科大学や埼玉県立大学と共同で行っているようであるため，その教員の一部を埼玉大学に移動したり，大学同士が合併したりすることを試みることによって，医学部や看護学部の設置を目指し，それをトリガとして，他大学との共同研究を推進・他の機関から他分野の教員を確保することによって，その他の学部学科を増設する事ができると良いと考えられる．
　学生数の確保が充実してくれば，さいたま市にも交渉がしやすくなると考えられる．新しく増設した学部学科の研究成果を見込んで，新たな共同研究・事業を行い，その知財や学問，技術をさいたま市に還元することを投資先とし，さいたま市と埼玉大学が双方にリターンを及ぼし合って発展することを目指すべきであると考える．
　以上のような施策を行うべきであると私は考える．以下には，施策の内容をまとめる．
- 埼玉医科大学や埼玉県立大学との連携強化
- 医学部や看護学部の設置あるいは埼玉医科大学・埼玉県立大学との合併
- 他大学との共同研究の推進・他の機関より他分野の教員の確保
- さいたま市への貢献を目的とした共同研究・事業の実行
- さいたま市による埼玉大学に対する資金的援助，研究施設の建設，他大学との共同研究マッチング

#### (3) 施策による効果の定量的推定
　前章で，埼玉大学の学部数は5個であり，千葉大学が10個，横浜国立大学が5個であるため，埼玉大学の学部数は少ないと述べた．しかし，正確には研究機関としての大学の学部は，大学院の学部数も含まれるため，埼玉大学は9個（＋4個），千葉大学は21個（＋11個），横浜国立大学は11個（＋6個）というのが正しい．この数に従って，科研費の合計額も変化していることが確認できる．学部数に対する科研費の平均値をプロットする．

```python
# 2次近似の計算
x = np.array([0, 9, 21, 11])
y = np.array([0, np.mean(df_su['fund']), np.mean(df_tu['fund']), np.mean(df_yu['fund'])])
coefficients = np.polyfit(x, y, 2)
linear_fit = np.poly1d(coefficients)
x_fit = np.linspace(min(x), max(x)+10, 100)
y_fit = linear_fit(x_fit)
print(f"Quadratic coefficient: {coefficients[0]}")
print(f"Linear coefficient: {coefficients[1]}")
print(f"Constant term: {coefficients[2]}")


#グラフの描画 
plt.figure()
plt.rcParams['font.family'] = 'Times New Roman'
plt.plot([9], df_su['fund'].mean(), label='Saitama University', marker='o')
plt.plot([21], df_tu['fund'].mean(), label='Chiba University', marker='o')
plt.plot([11], df_yu['fund'].mean(), label='Yokohama National University', marker='o')
plt.plot(x_fit, y_fit, 'r--', label='_nolegend_')
plt.xlabel('Number of Faculties')
plt.ylabel('Average of Grants-in-Aid for Scientific Research [1000 yen]')
plt.xlim(0, 25)  
plt.ylim(0, 5e6) 
plt.legend(bbox_to_anchor=(0.5, 1.2), loc='upper center', frameon=False, shadow=True, ncol=3, fancybox=False, edgecolor='black')
plt.grid(True)
plt.show()


# 13学部の場合の科研費予測を計算
faculties = 13
budget = coefficients[0] * faculties ** 2 + coefficients[1] * faculties + coefficients[2]
print(f'13 Faculties: {budget} [1000 yen]')
```

　以上より，原点を含む学部数と科研費の関係は，概ね2次関数で近似することができる．  
　もし，埼玉大学の学部と大学院にそれぞれ医学部と看護学部が増設されれば，合計で4学部増え，13学部となり，科研費の取得額は2312810千円となることが見込まれる．
　この資金を基に精力的に教育・研究を行った結果，少子高齢化するさいたま市へ医療・介護の人材が増えれば，さいたま市にとってもリターンとなるため，前節で述べたさいたま市と埼玉大学の関係が実現されやすくなると考えられる．したがって，埼玉医科大学および埼玉県立大学との交渉が重要になると私は考える．
